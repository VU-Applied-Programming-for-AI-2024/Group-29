```python
import json

import http.client

conn = http.client.HTTPSConnection("api.football-data.org")
headers = {
    'X-Auth-Token': 'e01ad631ff844c349b745cb14b38e770'
}

for team_id in range(80,115):
    conn.request("GET", f"/v4/teams/{team_id}", headers=headers)
    res = conn.getresponse()
    data = res.read()
    data_json = json.loads(data.decode("utf-8"))
    print(f"Team ID: {team_id}, Team Name: {data_json}")
```
